```python
from pybaseball import pitching_stats, playerid_reverse_lookup
import pandas as pd
```

```python
def get_pitcher_data(year, qual):
    pitcher_data = pitching_stats(year, qual=qual)

    pitcher_ids = pitcher_data["IDfg"].tolist()

    lookup_table = playerid_reverse_lookup(pitcher_ids, key_type="fangraphs")

    missing_ids = [id for id in pitcher_ids if id not in lookup_table["key_fangraphs"].values]

    trimmed_pitcher_data = pitcher_data[~pitcher_data["IDfg"].isin(missing_ids)]

    return lookup_table, trimmed_pitcher_data
    
```

```python
lookup_table, pitcher_data = get_pitcher_data(2024, 50)
```

```python
lookup_table.to_csv('collected_data/pitcher_lookup_table.csv', index=False)
```

```python
pitching_stat_categories = ['IDfg', 'ERA', 'GB%', 'WHIP', 'BABIP', 'FB%', 'SL%', 'CT%', 'CB%', 'CH%', 'SF%']
selected_pitching_stats = pitcher_data[pitching_stat_categories].fillna(0)
selected_pitching_stats.head()
```
